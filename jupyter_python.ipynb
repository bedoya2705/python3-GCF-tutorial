```python
model = Lmer('CH4cg ~ CMSC + PCO + (1 + CMSC + PCO|Animal)', data=tabred
display(model.fit())
# ANOVA results from fitted model
display(model.anova())
# Plot estimated model coefficients 
model.plot_summary()
```
